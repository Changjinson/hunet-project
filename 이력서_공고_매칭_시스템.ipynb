```python
def one_ten(label):
  if label == 10:
    return 1
  else:
    return label

def three_four(label):
    if label == 4:
      return 3
    else:
      return label

def nine_eleven(label):
    if label == 11:
      return 9
    else:
      return label

def thirteen_sixteen(label):
    if label == 16:
      return 11
    else:
      return label

data['label'] = data.label.apply(one_ten)
data['label'] = data.label.apply(three_four)
data['label'] = data.label.apply(nine_eleven)
data['label'] = data.label.apply(thirteen_sixteen)
data.head()
```

```python
import re

# 특수문자 제거
def clean_text(text):
  pattern = r'[^a-zA-Zㄱ-ㅎㅏ-ㅣ가-힣 ]'
  text_cleaned = re.sub(pattern, ' ', str(text))
  return text_cleaned
```

```python
from konlpy.tag import Mecab

# 불용어 제거
stop_words_korean = open('/content/drive/MyDrive/Colab Notebooks/hunet/saramain/stop_word_korean.txt', 'r', encoding='utf-8').readlines()
for i in range(len(stop_words_korean)):
  stop_words_korean[i]=stop_words_korean[i].replace('\n','')
stop_words_english = stopwords.words('english')
stop_words = stop_words_korean+stop_words_english

mecab = Mecab()

def del_stopwords(text, stop_words=stop_words):
  text_tokens = [x for x in mecab.morphs(text) if x not in stop_words and len(x) >= 2]
  return ' '.join(text_tokens)


data1['cont'] = data1['cont'].apply(del_stopwords)
data1.head()
```

# dataloader

```python
from torch.utils.data import Dataset
import torch

class job_data(Dataset):
  def __init__(self,df):
    self.input = df.sequence_ml
    self.label = df.label

  def __len__(self):
      return len(self.input)

  def __getitem__(self, index):
    return [
        torch.tensor(self.input[index]),
        torch.tensor([self.label[index]])
    ]

data1_sample = data1.sample(frac=1, random_state=111).reset_index(drop=True)

split_num = int(len(data1_sample)*0.8)
data1_train = data1_sample[:split_num]
data1_test = data1_sample[split_num:].reset_index(drop=True)

job_dataset_train = job_data(data1_train)
job_dataset_test = job_data(data1_test)
job_dataset_train[1][1]
```

```python
from torch.nn.utils.rnn import pad_sequence
from torch.utils.data import DataLoader

def collate_fn(batch):
  input = pad_sequence([b[0] for b in batch], batch_first=True)
  label = pad_sequence([b[1] for b in batch], batch_first=True)
  return input, label

train_dataloader = DataLoader(job_dataset_train, collate_fn=collate_fn, batch_size=32)
test_dataloader = DataLoader(job_dataset_test, collate_fn=collate_fn, batch_size=32)
```

# model

```python
import torch
from torch import nn

class classifier(nn.Module):
    def __init__(self, embedding_dim, model_dim, num_class, dropout_p, len_vocab):
        super(classifier, self).__init__()

        self.embedding_dim = embedding_dim
        self.model_dim = model_dim
        self.num_class = num_class
        self.dropout_p = dropout_p
        self.len_vocab = len_vocab

        self.embedding = torch.nn.Embedding(num_embeddings=self.len_vocab, embedding_dim=self.embedding_dim)
        self.bidirectional_gru_layer_em = torch.nn.GRU(input_size=self.embedding_dim, hidden_size=self.model_dim, 
                                                 batch_first=True, bidirectional=True)
        self.bidirectional_gru_layer = torch.nn.GRU(input_size=self.model_dim*2, hidden_size=self.model_dim, 
                                                 batch_first=True, bidirectional=True)
        
        self.classifier_layer = torch.nn.Linear(self.model_dim, self.num_class)
        self.softmax = torch.nn.Softmax(dim=-1)
        self.dropout = torch.nn.Dropout(self.dropout_p)


    def forward(self, inputs):
      embedding_inputs = self.embedding(inputs)
      gru = self.bidirectional_gru_layer_em(embedding_inputs)[0]
      gru = self.dropout(gru)
      gru1 = self.bidirectional_gru_layer(gru)[0]
      gru1 = self.dropout(gru1)
      gru2 = self.bidirectional_gru_layer(gru1)[1][0]
      gru2 = self.dropout(gru2)
      outputs = self.softmax(self.classifier_layer(gru2))
      return outputs
```

# train

```python
def train(model, train_dataloader, test_dataloader, loss_function, optimizer):
  model.train()
  total_loss = []
  for i, data in enumerate(train_dataloader):
    optimizer.zero_grad()
    
    inputs, labels = data
    inputs = inputs.to(device)
    labels = labels.to(device)
    outputs = model(inputs)
    labels = labels.squeeze()
    loss = loss_function(outputs, labels)
    total_loss.append(loss.item())
    loss.backward()
    optimizer.step()

  model.eval()
  test_losses = []
  with torch.no_grad():
    for data in test_dataloader:
      test_inputs, test_labels = data
      test_inputs = test_inputs.to(device)
      test_labels = test_labels.to(device)
      test_outputs = model(test_inputs)
      test_labels = test_labels.squeeze()
      test_loss = loss_function(test_outputs, test_labels)
      test_losses.append(test_loss.item())

  return sum(total_loss)/len(total_loss), sum(test_losses)/len(test_losses)
```

```python
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
print(device)
```

```python
embedding_dim = 512
model_dim = 512 
num_class = len(data1.label.value_counts())
dropout_p = 0.1
len_vocab = vocab_size
```

```python
classifier_model = classifier(embedding_dim, model_dim, num_class, dropout_p, len_vocab).to(device)
classifier_model.train()
```

```python
lr = 1e-4
epochs = 100
optimizer = torch.optim.Adam(classifier_model.parameters(), lr=lr)
loss_funtion = torch.nn.CrossEntropyLoss()
```

```python
cnt = 0
pre_test_loss = 100
train_losses = []
test_losses = []

for i in range(epochs):
  train_avg_loss, test_avg_loss = train(classifier_model, train_dataloader, test_dataloader, loss_funtion, optimizer)
  train_losses.append(train_avg_loss)
  test_losses.append(test_avg_loss)

  if pre_test_loss <= test_avg_loss: # test loss가 이전의 test loss보다 크면(잘못된 방향으로 업데이트된 경우)
    cnt += 1
  else: # test loss가 이전의 test loss보다 작다면(잘된 방향으로 업데이트된 경우)
    cnt = 0
    state_dict = classifier_model.state_dict()
    pre_test_loss = test_avg_loss

  print('epochs : ', i, ' average train loss : ',train_avg_loss, ' average test loss : ', test_avg_loss, 'cnt : ', cnt)
  if i >= 5 and cnt == 5:
    break
```

```python
torch.save(state_dict, '/content/drive/MyDrive/Colab Notebooks/hunet/saramain/gru_classifier.bin')
```

# inference

```python
pred = []
answers = []

classifier_model_load = classifier(embedding_dim, model_dim, num_class, dropout_p, len_vocab).to(device)
classifier_model_load.load_state_dict(torch.load('/content/drive/MyDrive/Colab Notebooks/hunet/saramain/gru_classifier.bin'))

classifier_model_load.eval()
with torch.no_grad():
  for data in test_dataloader:
    test_inputs, test_labels = data
    test_inputs = test_inputs.to(device)
    test_labels = test_labels.to(device)

    pred_labels = classifier_model_load(test_inputs).argmax(-1)
    answer = test_labels.squeeze()
    pred.extend(pred_labels.tolist())
    answers.extend(answer.tolist())

```

```python
from sklearn.metrics import confusion_matrix
import itertools

def plot_confusion_matrix(cm, classes,
                          normalize=True,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
  """
  This function prints and plots the confusion matrix.
  Normalization can be applied by setting `normalize=True`.
  """
  if normalize:
      cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
      print("Normalized confusion matrix")
  else:
      print('Confusion matrix, without normalization')

  print(cm)

  plt.imshow(cm, interpolation='nearest', cmap=cmap)
  plt.title(title)
  plt.colorbar()
  tick_marks = np.arange(len(classes))
  plt.xticks(tick_marks, classes, rotation=45)
  plt.yticks(tick_marks, classes)

  fmt = '.2f' if normalize else 'd'
  thresh = cm.max() / 2.
  for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
      plt.text(j, i, format(cm[i, j], fmt),
               horizontalalignment="center",
               color="white" if cm[i, j] > thresh else "black")

  plt.tight_layout()
  plt.ylabel('True label')
  plt.xlabel('Predicted label')
  plt.show()

plt.figure(figsize=(8,8))
classes = [x for x in range(num_class)]
cm = confusion_matrix(answers, pred)
plot_confusion_matrix(cm, classes)
```

```python
from sklearn.metrics import f1_score
f1_score(answers, pred, average='macro')
```

```python
import random
import pandas as pd

# 하나의 새로운 sequence를 만드는 함수
def sampling(df, k, max_length=512):
  index = random.randint(10, max_length)
  df_1 = df.loc[df.length==index]
  i = 1
  while True:
    df_2 = df.loc[df.length==index+i]
    df_3 = df.loc[df.length==index-i]
    df_1 = pd.concat([df_1, df_2, df_3], axis=0)
    i += 1
    if len(df_1) >= k:
      break
  elements = []
  for element in df_1.sequence_ml:
    elements.extend(element)
  new_sequences = []
  new_sequences_length = []
  for i in range(k):
    new_sequence = []
    for i in range(index):
      elem_index = random.randint(0, len(elements)-1)
      new_sequence.append(elements[elem_index])
    new_sequences.append(new_sequence)
    new_sequences_length.append(len(new_sequence))
  return new_sequences, new_sequences_length

# 하나의 label에 대하여 여러개의 sequence를 만드는 함수
def sampling_method(df, number, label):
  sequences = []
  labels = []
  lengths = []
  for i in range(number):
    new_sequence, length = sampling(df, 2)
    label_value = label
    sequences.append(new_sequence)
    labels.append(label_value)
    lengths.append(length)
  df_new = pd.DataFrame()
  df_new['sequence_ml'] = sequences
  df_new['label'] = labels
  df_new['length'] = lengths
  return df_new

# 전체 샘플링 코드
def new_sampling_method(df, num_dict):
  labels = list(data1.label.value_counts().index.values)
  for label in labels:
    num = num_dict[label]
    if num == 0:
      continue
    df_label = df.loc[df.label == label]
    df_sampled = sampling_method(df_label, num, label)
    df = pd.concat([df, df_sampled], axis=0)
  return df

num_dict = sample_dict = {0:0, 1:150, 2:150, 3:300, 4:0}

data_sampled = new_sampling_method(data1_train[['label', 'sequence_ml', 'length']], sample_dict)
len(data_sampled)
```

```python
from torch.nn.utils.rnn import pad_sequence
from torch.utils.data import DataLoader

def collate_fn(batch):
  input = pad_sequence([b[0] for b in batch], batch_first=True)
  label = pad_sequence([b[1] for b in batch], batch_first=True)
  return input, label

train_dataloader = DataLoader(job_dataset_train, collate_fn=collate_fn, batch_size=32)
```

```python
import torch
from torch import nn

class classifier(nn.Module):
    def __init__(self, embedding_dim, model_dim, num_class, dropout_p, len_vocab):
        super(classifier, self).__init__()

        self.embedding_dim = embedding_dim
        self.model_dim = model_dim
        self.num_class = num_class
        self.dropout_p = dropout_p
        self.len_vocab = len_vocab

        self.embedding = torch.nn.Embedding(num_embeddings=self.len_vocab, embedding_dim=self.embedding_dim)
        self.bidirectional_gru_layer_em = torch.nn.GRU(input_size=self.embedding_dim, hidden_size=self.model_dim, 
                                                 batch_first=True, bidirectional=True)
        self.bidirectional_gru_layer = torch.nn.GRU(input_size=self.model_dim*2, hidden_size=self.model_dim, 
                                                 batch_first=True, bidirectional=True)
        
        self.classifier_layer = torch.nn.Linear(self.model_dim, self.num_class)
        self.softmax = torch.nn.Softmax(dim=-1)
        self.dropout = torch.nn.Dropout(self.dropout_p)


    def forward(self, inputs):
      embedding_inputs = self.embedding(inputs)
      gru = self.bidirectional_gru_layer_em(embedding_inputs)[0]
      gru = self.dropout(gru)
      gru1 = self.bidirectional_gru_layer(gru)[0]
      gru1 = self.dropout(gru1)
      gru2 = self.bidirectional_gru_layer(gru1)[1][0]
      gru2 = self.dropout(gru2)
      outputs = self.softmax(self.classifier_layer(gru2))
      return outputs

        
        
```

```python
def train(model, train_dataloader, test_dataloader, loss_function, optimizer):
  model.train()
  total_loss = []
  for i, data in enumerate(train_dataloader):
    optimizer.zero_grad()
    
    inputs, labels = data
    inputs = inputs.to(device)
    labels = labels.to(device)
    outputs = model(inputs)
    labels = labels.squeeze()
    loss = loss_function(outputs, labels)
    total_loss.append(loss.item())
    loss.backward()
    optimizer.step()

  model.eval()
  test_losses = []
  with torch.no_grad():
    for data in test_dataloader:
      test_inputs, test_labels = data
      test_inputs = test_inputs.to(device)
      test_labels = test_labels.to(device)
      test_outputs = model(test_inputs)
      test_labels = test_labels.squeeze()
      test_loss = loss_function(test_outputs, test_labels)
      test_losses.append(test_loss.item())

  return sum(total_loss)/len(total_loss), sum(test_losses)/len(test_losses)
```
